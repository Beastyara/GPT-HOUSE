```python
import random
with open(r'C:\Users\User\Desktop\EXAMPLE.txt', 'w', encoding="utf-8") as file:
    for j in random.sample(range(len(df)), 20):#randomly get 20 samples 
        printingfeatures = ''
        features = ''
        for i in df.columns:
            if df.iloc[j][i] == 0:
                continue
            if df.iloc[j][i] == 'no': #Use only features those are existing 
                continue
            features = features + i + ': ' + str(df.iloc[j][i]) + '; '#Creating list of features for GPT
            printingfeatures = printingfeatures + i + ': ' + str(df.iloc[j][i]) + '\n'
        response = client.chat.completions.create(#run GPT for each sample
          model="gpt-3.5-turbo",
          messages=[
            {"role": "system", "content": "You are a Creative Instagram Post Writer."},
            {"role": "user", "content": "I will give you information that need to be in Instagram Post about selling the house. You need to use only information that is given and creatively express why feature is great for customers. Express only positive features."},
            {"role": "assistant", "content": "Ok, I will use only given information to write the creative post. Also, I will only express the positive features."},
            {"role": "user", "content": "Also, try to write not too long post. And do not add extra information or extra feature. Use only given features."}, 
            {"role": "assistant", "content": "Ok, I will try to write not too long post and I will use only given features and given information"}, 
            {"role": "user", "content": features},
          ],#It is starting dialog, also first sentence forces GPT to behave as Creative Instagram Post Writer
        )
        print(response.choices[0].message.content)
        file.write(printingfeatures)
        file.write('// all parameters used to generate this sample\n')
        file.write(response.choices[0].message.content)
        file.write('\n------------------------ // divider\n\n')
    
```
